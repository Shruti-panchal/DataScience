Document Tampering

purpose of this project is to detect tampering of Document using COMPUTER VISION. This project will help different organisation in detecting whether the id provided to them by their employees or customers or anyone is original or not.


For this project we will calculate the structural similarity of original Document and the Document uploaded by user.

```python
from skimage.metrics import structural_similarity
from skimage.transform import resize
import imutils
import cv2
from PIL import Image
import requests
```

```python
original = Image.open('./static/images/original/myadh.jpeg')
original
```

```python
tampered = Image.open('./static/images/uploaded/doc.jpeg')
tampered
```

```python
original.format, original.size
tampered.format, tampered.size
```

```python
original = original.resize((250, 160))
original.size
```

```python
tampered = tampered.resize((250, 160))
tampered.size
```

```python
original = cv2.imread('./static/images/original/myadh.jpeg')
tampered = cv2.imread('./static/images/uploaded/doc.jpeg')
```

```python
original_gray = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
original_gray.shape
```

```python
tampered_gray = cv2.cvtColor(tampered, cv2.COLOR_BGR2GRAY)
tampered_gray.shape
```

```python
if original_gray.shape != tampered_gray.shape:
    tampered_gray_resized = resize(tampered_gray, original_gray.shape, anti_aliasing=True)
    tampered_gray = tampered_gray_resized
else:
    tampered_gray_resized = tampered_gray

original_gray, tampered_gray
```

```python
# Compute the Structural Similarity Index (SSIM) between the two images, ensuring that the difference image is returned
(score, diff) = structural_similarity(original_gray, tampered_gray, full=True)
diff = (diff * 255).astype("uint8")
print("SSIM: {}".format(score))
```
